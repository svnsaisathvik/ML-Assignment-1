```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(train_df['Initial Health Score'], train_df['Recovery Index'], alpha=0.6, color='teal')
plt.title('Initial Health Score vs Recvery INdex')
plt.xlabel('Initial Health Score')
plt.ylabel('Recovery Index')
plt.grid(True)
plt.show()
```
